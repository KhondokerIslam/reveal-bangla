### Dataset: Preprocess (Not Needed Now)

```python
def process_reveal(df):
    result_list = []

    grouped = df.groupby('question_id')

    for question_id, group in grouped:
        group = group.sort_values('step_idx')

        processed_steps = []
        question_en = group['question_eng'].iloc[0]
        question_ben = group['question_ben'].iloc[0]
        evidence_eng = ""
        evidence_ben = ""
        step_eng = ""
        step_ben = ""

        max_step_idx = group['step_idx'].max()

        for _, row in group.iterrows():
            if row['step_idx'] == max_step_idx:
                continue

            if pd.notna(row['evidence_eng']) and row['evidence_eng'].strip():
                evidence_eng += f"{int(row['step_idx'])}. {row['evidence_eng']}\n"
                evidence_ben += f"{int(row['step_idx'])}. {row['evidence_ben']}\n"

            step_eng += f"{int(row['step_idx'])}. {row['step_eng']}\n"
            step_ben += f"{int(row['step_idx'])}. {row['step_ben']}\n"


        result_list.append({
            'question_id': question_id,
            'question_eng': question_en,
            'question_ben': question_ben,
            'evidence_eng': evidence_eng,
            'evidence_ben': evidence_ben,
            'step_eng': step_eng,
            'step_ben': step_ben,
        })

    return pd.DataFrame(result_list)
```

### Answer Generator Class

```python
class AnswerGenerator():

    def __init__(self, model_name):

        "Initializing `model` and default Jinja `chat_template`"

        self.model_name = model_name

        self.tokenizer = AutoTokenizer.from_pretrained(self.model_name)

        if model_name.split("/")[0] == "BanglaLLM":
            init_model = AutoPeftModelForCausalLM.from_pretrained(
                self.model_name,
                torch_dtype=torch.bfloat16
            )
            self.model = init_model.merge_and_unload()

        elif model_name.split("/")[0] == "meta-llama":
             self.model = LlamaForCausalLM.from_pretrained(
                self.model_name,
                torch_dtype=torch.bfloat16
            )

        self.chat_template = (
            '{{- bos_token }}\n'
            '{#- This block extracts the system message, so we can slot it into the right place. #}\n'
            "{%- if messages[0]['role'] == 'system' %}\n"
            "    {%- set system_message = messages[0]['content']|trim %}\n"
            '    {%- set messages = messages[1:] %}\n'
            '{%- else %}\n'
            '    {%- set system_message = "" %}\n'
            '{%- endif %}\n'
            '\n'
            '{#- System message #}\n'
            '{{- "<|start_header_id|>system<|end_header_id|>\\n\\n" }}\n'
            '{{- system_message }}\n'
            '{{- "<|eot_id|>" }}\n'
            '\n'
            '{%- for message in messages %}\n'
            "        {{- '<|start_header_id|>' + message['role'] + '<|end_header_id|>\\n\\n'+ message['content'] | trim + '<|eot_id|>' }}\n"
            '{%- endfor %}\n'
            '{%- if add_generation_prompt %}\n'
            "    {{- '<|start_header_id|>assistant<|end_header_id|>\\n\\n' }}\n"
            '{%- endif %}\n'
        )

    def int_seed( value = 0. ):
        torch.manual_seed(value)

    def load_apply_tokenizer( self, chat, continue_final_message=True ):

        self.tokenizer.pad_token_id = self.tokenizer.eos_token_id

        self.tokenizer.chat_template = self.chat_template

        tokenized_chat = self.tokenizer.apply_chat_template(
            chat,
            tokenize=True,
            continue_final_message=continue_final_message,
            return_tensors="pt",
            padding=True,
            padding_side="left"
        )

        print( f"**Tokenized Chat: \n{tokenized_chat}\n******")

        # un_tokenized_chat = self.tokenizer.apply_chat_template(
        #     chat,
        #     tokenize=False,
        #     continue_final_message=continue_final_message,
        #     return_tensors="pt",
        #     padding=True,
        #     padding_side="left"
        # )
        # print( f"**UnTokenized Chat: \n{un_tokenized_chat}\n******")

        self.input_ids = tokenized_chat
        self.attention_mask = (self.input_ids != self.tokenizer.pad_token_id).long()
        # print(self.tokenizer.decode(tokenized_chat[0]))

    def generate_w_model(self, do_sample = False, temperature=None, top_p=None, top_k=None, max_new_tokens=256):

        self.model.eval()
        outputs = self.model.generate(
            input_ids=self.input_ids,
            attention_mask=self.attention_mask,
            pad_token_id=self.tokenizer.pad_token_id,
            do_sample=do_sample,
            temperature=temperature,
            top_p=top_p,
            top_k=top_k,
            max_new_tokens=max_new_tokens
        )

        # print("Output:\n" + 100 * '-')
        # print(self.tokenizer.decode(outputs[0], skip_special_tokens=True))
        # print("" + 100 * '-')

        model_gen_resp = html.unescape(self.tokenizer.decode(outputs[0], skip_special_tokens=True))
        model_special_token_gen_resp = html.unescape(self.tokenizer.decode(outputs[0], skip_special_tokens=False))

        return model_gen_resp, model_special_token_gen_resp
```

### Push To HuggingFace

```python
def result_push_hf( data, filename ):

    # print(f"data: {data}")

    # Convert dictionary to JSON string and encode it to bytes
    json_str = json.dumps(data, ensure_ascii=False, indent=4)
    json_bytes = BytesIO(json_str.encode('utf-8'))  # Encode string to bytes
    json_bytes.seek(0)  # Reset buffer position to the beginning

    # print(f"json_str: {json_str}")

    # Hugging Face repo details
    repo_id = "khondoker/results-reveal-bangla"

    # Upload to Hugging Face without saving locally
    api = HfApi()
    api.upload_file(
        path_or_fileobj=json_bytes,  # In-memory file object
        path_in_repo=filename,   # Destination in the repo
        repo_id=repo_id,            # Your repository ID
        repo_type="dataset"         # Can be 'dataset' or 'model'
    )
```
